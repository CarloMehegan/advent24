```python
#create a 2d matrix with the input
ws = []

with open('day4.txt', 'r') as file:
    for line in file:
        ws.append(list(line))

print(ws)
```

```python
def search_direction(direction):
    #ill be real i made copilot write this table
    if direction == "up":
        return (0, -1)
    elif direction == "down":
        return (0, 1)
    elif direction == "left":
        return (-1, 0)
    elif direction == "right":
        return (1, 0)
    elif direction == "upleft":
        return (-1, -1)
    elif direction == "upright":
        return (1, -1)
    elif direction == "downleft":
        return (-1, 1)
    elif direction == "downright":
        return (1, 1)

DIRECTIONS = ["up", "down", "left", "right", "upleft", "upright", "downleft", "downright"]

def is_in_range(x,y):
    return x >= 0 and x < 140 and y >= 0 and y < 140

def search_for_match(x, y):
    found = 0
    if ws[y][x] == "X": #always start with X
        for direction in DIRECTIONS:
            dx, dy = search_direction(direction)
            #brute
            if is_in_range(x+dx+dx+dx, y+dy+dy+dy): #check the next three are in range
                if ws[y+dy][x+dx] == "M":
                    if ws[y+dy+dy][x+dx+dx] == "A":
                        if ws[y+dy+dy+dy][x+dx+dx+dx] == "S":
                            #found a match!
                            found += 1
    return found


total = 0

for y in range(len(ws)):
    for x in range(len(ws[y])):
        total += search_for_match(x,y)

total
```

now we are looking for M-A-S crosses

we can start with A's and look in the diagonals for M's and S's

if there are two M's and two S's, and the M's and S's are opposite(?) each other, then we match

```python
def search_direction(direction):
    if direction == "up":
        return (0, -1)
    elif direction == "down":
        return (0, 1)
    elif direction == "left":
        return (-1, 0)
    elif direction == "right":
        return (1, 0)
    elif direction == "upleft":
        return (-1, -1)
    elif direction == "upright":
        return (1, -1)
    elif direction == "downleft":
        return (-1, 1)
    elif direction == "downright":
        return (1, 1)

#change DIRECTIONS to only include diagonals
DIRECTIONS = ["upleft", "upright", "downleft", "downright"]

def is_in_range(x,y):
    return x >= 0 and x < 140 and y >= 0 and y < 140

def search_for_match(x, y):
    found = 0
    if ws[y][x] == "A": #always start with A
        cross_matches = []
        for direction in DIRECTIONS:
            dx, dy = search_direction(direction)
            if is_in_range(x+dx, y+dy): #check the next three are in range
                if ws[y+dy][x+dx] == "M" or ws[y+dy][x+dx] == "S":
                    cross_matches.append(ws[y+dy][x+dx])
        
        #now let's look at cross matches.
        #correct configs, since we go upleft then upright then downleft then downright:
        #MMSS
        #MSMS
        #SSMM
        #SMSM
        #anything else is bad
        return cross_matches in [["M", "M", "S", "S"], ["M", "S", "M", "S"], ["S", "S", "M", "M"], ["S", "M", "S", "M"]]

    return False


total = 0

for y in range(len(ws)):
    for x in range(len(ws[y])):
        total += search_for_match(x,y)

total
```
