# red-nosed reports

step 1. parse each line as a list

step 2. go through the list and check if the first two numbers have a positive or negative difference

step 3. also check if that difference is in the correct range

step 4. keep iterating to see if we change from positive to negative or vice versa, or if difference is out of range

is there a efficient way to do it? what if when we parse the data and save the lists, we just save the differences instead of the numbers? that might be bad for part 2

final answer = how many reports are safe

```python
safe_reports = []

with open('day2.txt', 'r') as file:
    #for each line, iterate through the numbers and create a list of the differences between each consecutive number
    for line in file:
        numbers = [int(x) for x in line.split()]
        differences = [numbers[i] - numbers[i+1] for i in range(len(numbers)-1)]

        #we can just check if they're safe right here
        #a new python function i just learned, all()
        correct_range = all(1 <= abs(diff) <= 3 for diff in differences)

        #we can use all() for this check too
        #all is efficient because it stops checking as soon as it finds a false
        all_positive = all(diff > 0 for diff in differences)
        all_negative = all(diff < 0 for diff in differences)

        if (all_positive or all_negative) and correct_range:
            safe_reports.append(differences)

safe_reports
```
